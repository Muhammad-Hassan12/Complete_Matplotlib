## Specifying Legends
We can also specify Legends! But, label is necessary!

```python
x= np.arange(50,1050,20)
y= np.tan(x)
plt.plot(x,y,label='Tangent of "X"')
plt.legend()
```

## Grid-Lines
We can also use grid-lines!

```python
x= np.arange(50,1050,20)
y= np.tan(x)
plt.plot(x,y,label='Tangent of "X"')
plt.legend()
plt.grid(True, which='both') # "True" is for allowing grid and "which" represents axis and some other features!
```

#### Grid-Lines Color
We can also specify a color for grid-lines! (It also supports RGB as-well)

```python
x= np.arange(50,1050,20)
y= np.tan(x)
plt.plot(x,y,label='Tangent of "X"')
plt.legend()
plt.grid(True, which='both', color='magenta')
```

#### Minor-Ticks
We can also use "Minor-Grids a.k.a "Minor-ticks" but we have to wnable it sice this function is disable by default in "matplotlib"!

```python
x= np.arange(50,1050,20)
y= np.tan(x)
plt.plot(x,y,label='Tangent of "X"')
plt.legend()
plt.grid(True, which='major', color='magenta',linestyle="-")
plt.grid(True, which='minor', color='green',linestyle="--")
plt.minorticks_on() # This command is required for enabeling minor-ticks!
```

## 3) Bar-Digram or Bar-Graph

```python
x = ['Warren B','Larry E','Mark Z','Elon M','Bill G','Larry P']   # Subject for Bar-Graph!
Net_Worth = [119.1,144.2,110.9,236.1,117.5,108.6]   # Numbers for Bar-Graph!
x_position = [0,1,2,3,4,5]   # Position for Bar-Graph!
plt.bar(x_position, Net_Worth, color='red')   # Now plotting the graph by its position, Numbers and a desiered color!
plt.xlabel('Billioners')
plt.ylabel('Net Worth')
plt.title("World's richest Person (8-5-2023)")
plt.xticks(x_position, x)   # Setting up the ticks on x-axis!
plt.show()
```

## 6) Pie-Plotting
In the 4th step we are plotting "Pie-graph" on "ax1" you can also do this by "plt(matplotlib)". Now in the parenthesis we have "data" in which we have assigned the values for the plot, "Labels" are assigned with subject, "autopct" is showing the pie-chart in a percentage distribution, "Shadow" is adding shadows in the graph (it is not a necessary perk!), "startangle" starts the graph with the given angle (You can put a desired number of angle)!

```python
Subject = ['Warren B','Larry E','Mark Z','Elon M','Bill G','Larry P']  # Subject Of the plot!
data = [119.1,144.2,110.9,236.1,117.5,108.6]  # Data for the graph!
fig1, ax1 = plt.subplots()  # "Fig1" is a name of figure, and "ax1" is the name of axis, ther name can be anything!
ax1.pie(data, labels=Subject, autopct='%1.1f%%', shadow=True, startangle=90)  # The explaination of this step is given above!
ax1.axis('equal')  # All axis are equal!
plt.show()
```

## Saving Figures
We can also save figures as a "png" file! Also I am using the graph that I have made earlier!

```python
Subject = ['Warren B','Larry E','Mark Z','Elon M','Bill G','Larry P']
data = [119.1,144.2,110.9,236.1,117.5,108.6]
fig1, ax1 = plt.subplots()
ax1.pie(data, labels=Subject, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
fig1.savefig("Practice")
```

## Sub-Plotting
###### Definition:
Having more than one plot in a figure!

```python
x = np.linspace(0.1, 3**np.pi ,400)
y = np.sin(x*2*8)
f, (ax1, ax2) = plt.subplots(1,2)   # Two Figures in one row!
ax1.plot(x, y)
ax1.set_title('Figure 1')   # Title of first figure!
ax2.scatter(x, y)
ax2.set_title('Figure 2')   # Title of second figure!
```
